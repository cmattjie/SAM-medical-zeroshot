```python
import os
import numpy as np
from PIL import Image

# Replace 'input_folder' with the path to the folder containing the original masks
original_folder = '/mnt/B-SSD/maltamed/datasets/2D/hip/mask'

# Create output folders
femur_folder = '/mnt/B-SSD/maltamed/datasets/2D/hip/femur'
ilios_folder = '/mnt/B-SSD/maltamed/datasets/2D/hip/ilios'

# Create the new folders for the femur and ilios masks
os.makedirs(femur_folder, exist_ok=True)
os.makedirs(ilios_folder, exist_ok=True)

# Iterate through each file in the original folder
for filename in os.listdir(original_folder):
    if filename.endswith('.png') or filename.endswith('.jpg'):
        # Load the original mask image
        mask_path = os.path.join(original_folder, filename)
        mask_image = Image.open(mask_path)
        mask_array = np.array(mask_image)

        # Create the femur mask (background and femur)
        femur_mask_array = np.where(mask_array == 127, 255, 0).astype(np.uint8)
        femur_mask_image = Image.fromarray(femur_mask_array)
        femur_mask_path = os.path.join(femur_folder, filename)
        femur_mask_image.save(femur_mask_path)

        # Create the ilios mask (background and ilios)
        ilios_mask_array = np.where(mask_array == 255, 255, 0).astype(np.uint8)
        ilios_mask_image = Image.fromarray(ilios_mask_array)
        ilios_mask_path = os.path.join(ilios_folder, filename)
        ilios_mask_image.save(ilios_mask_path)

        print(f'Processed {filename}')

print('Done!')
```
